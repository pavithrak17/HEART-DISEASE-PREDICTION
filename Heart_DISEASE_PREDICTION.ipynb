```python
LoR=LogisticRegression()
```

```python
print(type(LoR))
```

```python
LoR.fit(X_train,y_train)
```

```python
y_pred=LoR.predict(X_test)
y
```

```python
from sklearn.metrics import accuracy_score
```

```python
asc=accuracy_score(y_test,y_pred)
asc
```
